````python
import numpy as np

a = np.array([1, 2, 3])   # Create a rank 1 array
print(type(a))            # Prints "<class 'numpy.ndarray'>"
print(a.shape)            # Prints "(3,)"
print(a[0], a[1], a[2])   # Prints "1 2 3"
a[0] = 5                  # Change an element of the array
print(a)                  # Prints "[5, 2, 3]"

b = np.array([[1,2,3],[4,5,6]])    # Create a rank 2 array
print(b.shape)                     # Prints "(2, 3)"
print(b[0, 0], b[0, 1], b[1, 0])   # Prints "1 2 4"
````

A slice of an array is a view into the same data, so modifying it will modify the original array.
```

```

````python
# 加減乘除
import numpy as np

x = np.array([[1,2],[3,4]], dtype=np.float64)
y = np.array([[5,6],[7,8]], dtype=np.float64)

# add
print(x+y)
print(np.add(x,y))

# subtract
print(x-y)
print(np.subtract(x,y))

# product
print(x*y)
print(np.multiply(x,y))

# division
print(x / y)
print(np.divide(x, y))
````

# Using Matplotlib to draw sine function

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
````

draw sine and cosine at once

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
ysin = np.sin(x)
ycos = np.cos(x)

plt.plot(x,ysin)
plt.plot(x,ycos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('sine and cosine')
plt.legend(['sin','cos'])
plt.show()
````

# Separate plot into subpolt

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3*np.pi, 0.1)
ySin = np.sin(x)
yCos = np.cos(x)

# height 2 and width 1, and it's first plot
plt.subplot(2,1,1)
plt.plot(x,ySin)
plt.title('sine')

# height 2 and width 1, and it's second plot
plt.subplot(2,1,2)
plt.plot(x,yCos)
plt.title('cosine')

plt.show()
````
