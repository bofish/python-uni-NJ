聚類分析 Cluster
===

利用K-mean function將數據分類

例子，算出學霸組與非學霸組：


````python
from pylab import *
from scipy.cluster.vq import *
list1 = [88.0,64.0,96.0,85.0]
list2 = [92.0,99.0,95.0,94.0]
list3 = [91.0,87.0,99.0,95.0]
list4 = [78.0,99.0,97.0,81.0]
list5 = [88.0,78.0,98.0,84.0]
list6 = [100.0,95.0,100.0,92.0]
data = vstack((list1,list2,list3,list4,list5,list6))
# the value type in list must be double or float.
centroids,_ =kmeans(data,2)
result,_ = vq(data,centroids)
print(result)
````

Matplotlib 與 Pylib 繪圖
===

Matplotlob.pypolt - 繪圖API，與MATLAB相似

Pylib - 集成庫（包含NumPy與PyPlot常用的函數）

### 利用Matplotlib繪折線圖:


````python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0,1,0.1)
plt.plot(t,t,t,t**2,t,t**3)
plt.show()
````

### 繪散點圖 - 加上 'o'

````python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0,1,0.1)
plt.plot(t,t**2,'o')
plt.show()
````

### 繪直方圖 - 將plot改為bar

````python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0,1,0.1)
plt.bar(t,t)
plt.show()
````

### 利用pylab繪製

````python
import numpy as np
import pylab as pl

t=np.arange(0,1,0.1)
pl.plot(t,t,t,t**2,t,t**3);
pl.show()
````
